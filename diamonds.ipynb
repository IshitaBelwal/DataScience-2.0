## Dropping rows that have 0 as one of the dimensions

```python
a.drop(a[a['x']==0].index,inplace=True)
a.drop(a[a['y']==0].index,inplace=True)
a.drop(a[a['z']==0].index,inplace=True)
```

```python
a.describe()
```

## Linear Regression Model fit between numerical features and target feature

```python
plt.figure(figsize=(15,5))
sns.regplot(x=a.price,y=a.carat)
plt.title("Price vs Carat")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.regplot(x=a.price,y=a.depth,color='red')
plt.title("Price vs Depth")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.regplot(x=a.price,y=a.table,color='green')
plt.title("Price vs Table")
plt.show()
```

## Box Plots and Density plots to analyze outliers

```python
plt.figure(figsize=(15,5),dpi=100)
sns.distplot(a.depth,kde=True)
plt.title('Depth Density Plot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(a.depth)
plt.title('Depth Boxplot')
plt.show()
```

```python
plt.figure(figsize=(15,6))

sns.displot(a.table,kde=True)
plt.title('Table Depth density plot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(a.table)
plt.title('Table Boxplot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(a.x)
plt.title('x dimension Boxplot')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(a.y)
plt.title('y dimension Boxplot')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(a.z)
plt.title('z dimension Boxplot')
plt.show()
```
